## Step 1: Generate Comprehensive Dataset in Kotlin

Let's create a rich dataset that demonstrates various visualization scenarios.

````kotlin
%js
// Import data from Kotlin
import { monthlySales } from '@jupyter';
import { Chart } from 'chart.js/auto';

// Get the cell container and render user profile
const container = getContainer();
container.innerHTML = `
    <div style="font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; max-width: 800px; margin: 0 auto;">
        <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); color: white; padding: 20px; border-radius: 12px; margin-bottom: 20px; text-align: center;">
            <h1 style="margin: 0; font-size: 24px;">ðŸ“Š Chart.js Bar Charts</h1>
            <p style="margin: 8px 0 0 0; opacity: 0.9;">Monthly Sales Performance Analysis</p>
        </div>
        <canvas id="barChart" style="max-width: 100%; height: 400px;"></canvas>
    </div>
`;

// Render bar chart
const barCtx = document.getElementById('barChart');
new Chart(barCtx, {
    type: 'bar',
    data: {
        labels: Object.keys(monthlySales),
        datasets: [{
            label: 'Monthly Sales ($)',
            data: Object.values(monthlySales),
            backgroundColor: 'rgba(102, 126, 234, 0.6)',
            borderColor: 'rgba(102, 126, 234, 1)',
            borderWidth: 2,
            borderRadius: 8
        }]
    },
    options: {
        responsive: true,
        maintainAspectRatio: false,
        plugins: {
            legend: {
                display: true,
                position: 'top'
            },
            tooltip: {
                callbacks: {
                    label: (context) => `Sales: $${context.parsed.y.toLocaleString()}`
                }
            }
        },
        scales: {
            y: {
                beginAtZero: true,
                ticks: {
                    callback: (value) => '$' + value.toLocaleString()
                }
            }
        },
        animation: {
            duration: 1500,
            easing: 'easeInOutQuart'
        }
    }
});
````

## Step 2: Chart.js - Bar Charts

Perfect for comparing values across categories and showing trends over time.

````kotlin
%js
// Import product trends data
import { productTrends } from '@jupyter';
import { Chart } from 'chart.js/auto';

// Get the cell container and render user profile
const container = getContainer();
container.innerHTML = `
    <div style="font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; max-width: 800px; margin: 0 auto;">
        <div style="background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); color: white; padding: 20px; border-radius: 12px; margin-bottom: 20px; text-align: center;">
            <h1 style="margin: 0; font-size: 24px;">ðŸ“ˆ Chart.js Line Charts</h1>
            <p style="margin: 8px 0 0 0; opacity: 0.9;">Multi-Series Product Trends Analysis</p>
        </div>
        <canvas id="lineChart" style="max-width: 100%; height: 400px;"></canvas>
    </div>
`;

// Prepare data for line chart
const months = productTrends.map(d => d.month);
const products = ['laptops', 'phones', 'tablets', 'accessories'];
const colors = ['#667eea', '#f093fb', '#4facfe', '#43e97b'];

// Render line chart
const lineCtx = document.getElementById('lineChart');
new Chart(lineCtx, {
    type: 'line',
    data: {
        labels: months,
        datasets: products.map((product, index) => ({
            label: product.charAt(0).toUpperCase() + product.slice(1),
            data: productTrends.map(d => d[product]),
            borderColor: colors[index],
            backgroundColor: colors[index] + '20',
            borderWidth: 3,
            tension: 0.4,
            fill: true,
            pointRadius: 5,
            pointHoverRadius: 7
        }))
    },
    options: {
        responsive: true,
        maintainAspectRatio: false,
        plugins: {
            legend: {
                display: true,
                position: 'top'
            },
            tooltip: {
                mode: 'index',
                intersect: false
            }
        },
        scales: {
            y: {
                beginAtZero: true,
                title: {
                    display: true,
                    text: 'Sales Volume'
                }
            }
        },
        animation: {
            duration: 2000,
            easing: 'easeInOutQuart'
        }
    }
});
````

## Step 3: Chart.js - Line Charts

Ideal for showing trends and changes over time with smooth curves.

````kotlin
%js
// Import category performance data
import { categoryPerformance, satisfactionData } from '@jupyter';
import { Chart } from 'chart.js/auto';

// Get the cell container and render user profile
const container = getContainer();
container.innerHTML = `
    <div style="font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; max-width: 900px; margin: 0 auto;">
        <div style="background: linear-gradient(135deg, #fa709a 0%, #fee140 100%); color: white; padding: 20px; border-radius: 12px; margin-bottom: 20px; text-align: center;">
            <h1 style="margin: 0; font-size: 24px;">ðŸ¥§ Chart.js Pie & Doughnut Charts</h1>
            <p style="margin: 8px 0 0 0; opacity: 0.9;">Proportion Analysis and Customer Satisfaction</p>
        </div>
        <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 20px;">
            <div>
                <canvas id="pieChart" style="max-width: 100%; height: 350px;"></canvas>
            </div>
            <div>
                <canvas id="doughnutChart" style="max-width: 100%; height: 350px;"></canvas>
            </div>
        </div>
    </div>
`;

// Pie chart colors
const pieColors = ['#667eea', '#f093fb', '#4facfe', '#43e97b', '#fa709a'];

// Render pie chart (Category Performance)
const pieCtx = document.getElementById('pieChart');
new Chart(pieCtx, {
    type: 'pie',
    data: {
        labels: Object.keys(categoryPerformance),
        datasets: [{
            data: Object.values(categoryPerformance),
            backgroundColor: pieColors,
            borderWidth: 2,
            borderColor: '#fff'
        }]
    },
    options: {
        responsive: true,
        maintainAspectRatio: false,
        plugins: {
            legend: {
                position: 'bottom'
            },
            title: {
                display: true,
                text: 'Category Performance (%)',
                font: { size: 16 }
            },
            tooltip: {
                callbacks: {
                    label: (context) => `${context.label}: ${context.parsed}%`
                }
            }
        },
        animation: {
            animateRotate: true,
            animateScale: true,
            duration: 1500
        }
    }
});

// Doughnut chart colors
const doughnutColors = ['#10b981', '#3b82f6', '#8b5cf6', '#f59e0b', '#ef4444'];

// Render doughnut chart (Customer Satisfaction)
const doughnutCtx = document.getElementById('doughnutChart');
new Chart(doughnutCtx, {
    type: 'doughnut',
    data: {
        labels: Object.keys(satisfactionData),
        datasets: [{
            data: Object.values(satisfactionData),
            backgroundColor: doughnutColors,
            borderWidth: 2,
            borderColor: '#fff'
        }]
    },
    options: {
        responsive: true,
        maintainAspectRatio: false,
        plugins: {
            legend: {
                position: 'bottom'
            },
            title: {
                display: true,
                text: 'Customer Satisfaction (%)',
                font: { size: 16 }
            },
            tooltip: {
                callbacks: {
                    label: (context) => `${context.label}: ${context.parsed}%`
                }
            }
        },
        cutout: '60%',
        animation: {
            animateRotate: true,
            animateScale: true,
            duration: 1500
        }
    }
});
````

## Step 4: Chart.js - Pie & Doughnut Charts

Perfect for showing proportions and percentages of a whole.

````kotlin
%js
// Import regional data
import { regionalData } from '@jupyter';
import * as echarts from 'echarts';

// Get the cell container and render user profile
const container = getContainer();
container.innerHTML = `
    <div style="font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; max-width: 800px; margin: 0 auto;">
        <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); color: white; padding: 20px; border-radius: 12px; margin-bottom: 20px; text-align: center;">
            <h1 style="margin: 0; font-size: 24px;">ðŸŽ¨ ECharts Advanced Visualizations</h1>
            <p style="margin: 8px 0 0 0; opacity: 0.9;">Regional Sales Performance with Stacked Bar Chart</p>
        </div>
        <div id="echartsContainer" style="width: 100%; height: 500px;"></div>
    </div>
`;

// Initialize ECharts
const myChart = echarts.init(document.getElementById('echartsContainer'));

// Prepare data
const regions = regionalData.map(d => d.region);
const q1Data = regionalData.map(d => d.q1);
const q2Data = regionalData.map(d => d.q2);
const q3Data = regionalData.map(d => d.q3);
const q4Data = regionalData.map(d => d.q4);

// Set chart options
const option = {
    title: {
        text: 'Regional Quarterly Sales',
        left: 'center',
        textStyle: {
            fontSize: 20,
            fontWeight: 'bold'
        }
    },
    tooltip: {
        trigger: 'axis',
        axisPointer: {
            type: 'shadow'
        },
        formatter: (params) => {
            let result = params[0].name + '<br/>';
            params.forEach(item => {
                result += `${item.marker} ${item.seriesName}: $${item.value.toLocaleString()}<br/>`;
            });
            return result;
        }
    },
    legend: {
        data: ['Q1', 'Q2', 'Q3', 'Q4'],
        bottom: 10
    },
    grid: {
        left: '3%',
        right: '4%',
        bottom: '15%',
        containLabel: true
    },
    xAxis: {
        type: 'category',
        data: regions
    },
    yAxis: {
        type: 'value',
        axisLabel: {
            formatter: (value) => '$' + (value / 1000) + 'K'
        }
    },
    series: [
        {
            name: 'Q1',
            type: 'bar',
            stack: 'total',
            data: q1Data,
            itemStyle: { color: '#5470c6' }
        },
        {
            name: 'Q2',
            type: 'bar',
            stack: 'total',
            data: q2Data,
            itemStyle: { color: '#91cc75' }
        },
        {
            name: 'Q3',
            type: 'bar',
            stack: 'total',
            data: q3Data,
            itemStyle: { color: '#fac858' }
        },
        {
            name: 'Q4',
            type: 'bar',
            stack: 'total',
            data: q4Data,
            itemStyle: { color: '#ee6666' }
        }
    ],
    toolbox: {
        feature: {
            saveAsImage: { title: 'Save' },
            dataView: { title: 'Data View', readOnly: false },
            magicType: { type: ['line', 'bar'], title: { line: 'Line', bar: 'Bar' } }
        }
    },
    animationDuration: 1500
};

myChart.setOption(option);
````

## Step 5: ECharts - Advanced Visualizations

ECharts offers more sophisticated chart types and built-in interactions.

````kotlin
%js
// Import and prepare radar chart data
import { monthlySales, categoryPerformance, satisfactionData } from '@jupyter';
import * as echarts from 'echarts';

// Get the cell container and render user profile
const container = getContainer();
container.innerHTML = `
    <div style="font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; max-width: 800px; margin: 0 auto;">
        <div style="background: linear-gradient(135deg, #ffecd2 0%, #fcb69f 100%); color: #333; padding: 20px; border-radius: 12px; margin-bottom: 20px; text-align: center;">
            <h1 style="margin: 0; font-size: 24px;">ðŸŽ¯ ECharts Interactive Radar Chart</h1>
            <p style="margin: 8px 0 0 0; opacity: 0.9;">Multi-Dimensional Performance Analysis</p>
        </div>
        <div id="radarChart" style="width: 100%; height: 500px;"></div>
    </div>
`;

// Initialize radar chart
const radarChart = echarts.init(document.getElementById('radarChart'));

// Prepare radar data
const radarOption = {
    title: {
        text: 'Business Performance Radar',
        left: 'center',
        textStyle: {
            fontSize: 20,
            fontWeight: 'bold'
        }
    },
    tooltip: {
        trigger: 'item'
    },
    legend: {
        data: ['Current Quarter', 'Target'],
        bottom: 10
    },
    radar: {
        indicator: [
            { name: 'Sales Volume', max: 100 },
            { name: 'Customer Satisfaction', max: 100 },
            { name: 'Market Share', max: 100 },
            { name: 'Product Quality', max: 100 },
            { name: 'Innovation', max: 100 },
            { name: 'Team Performance', max: 100 }
        ],
        radius: '60%',
        splitNumber: 4,
        axisName: {
            color: '#333',
            fontSize: 12
        },
        splitArea: {
            areaStyle: {
                color: ['rgba(250, 112, 154, 0.1)', 'rgba(250, 112, 154, 0.2)',
                        'rgba(250, 112, 154, 0.3)', 'rgba(250, 112, 154, 0.4)']
            }
        }
    },
    series: [{
        name: 'Performance Metrics',
        type: 'radar',
        data: [
            {
                value: [78, 85, 72, 88, 65, 82],
                name: 'Current Quarter',
                areaStyle: {
                    color: 'rgba(102, 126, 234, 0.3)'
                },
                lineStyle: {
                    color: '#667eea',
                    width: 2
                }
            },
            {
                value: [90, 90, 85, 95, 80, 90],
                name: 'Target',
                areaStyle: {
                    color: 'rgba(240, 147, 251, 0.2)'
                },
                lineStyle: {
                    color: '#f093fb',
                    width: 2,
                    type: 'dashed'
                }
            }
        ]
    }],
    toolbox: {
        feature: {
            saveAsImage: { title: 'Save' }
        }
    },
    animationDuration: 2000
};

radarChart.setOption(radarOption);
````
