This example demonstrates how to use the jsConfig DSL to configure JavaScript libraries in Kotlin Jupyter.

## Features
- Library configuration using DSL
- Preset library loading
- Custom library mapping
- Configuration management


```kotlin
// Basic jsConfig DSL usage
jsConfig {
    // Load individual libraries using DSL methods
    echart()      // ECharts for charts
    d3()          // D3.js for custom visualizations
    react()       // React for UI components
    lodash()      // Lodash for utilities
    moment()      // Moment.js for date handling
    axios()       // Axios for HTTP requests
}

println("Basic libraries configured successfully!")
```

```kotlin
// Advanced jsConfig DSL usage with custom libraries
jsConfig {
    // Load preset libraries
    echart()
    react()
    lodash()
    
    // Add custom library mappings
    mapLibrary("custom-lib", "https://cdn.example.com/custom-lib@1.0.0/dist/custom-lib.min.js")
    
    // Add library with dependencies
    mapLibrary("complex-lib", "https://cdn.example.com/complex-lib@2.0.0/dist/main.js", 
               listOf("https://cdn.example.com/complex-lib@2.0.0/dist/dependency.js"))
    
    // Add library with custom name
    mapLibrary("myChart", "https://cdn.jsdelivr.net/npm/chart.js@4.2.1/dist/chart.min.js")
}

println("Advanced configuration completed!")
```
