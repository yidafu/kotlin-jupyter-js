This example demonstrates how to use ECharts library in Kotlin Jupyter to create various types of charts.

## Features
- Line charts
- Bar charts  
- Pie charts
- Data visualization
- Interactive charts


```python
// Configure ECharts library
jsConfig {
    echart()
}

// Prepare sample data for different chart types
val salesData = listOf(120, 200, 150, 80, 70, 110, 130)
val months = listOf("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul")

val productSales = mapOf(
    "Laptops" to 45,
    "Phones" to 30,
    "Tablets" to 15,
    "Accessories" to 10
)

val quarterlyRevenue = listOf(
    mapOf("quarter" to "Q1", "revenue" to 120000, "profit" to 30000),
    mapOf("quarter" to "Q2", "revenue" to 150000, "profit" to 45000),
    mapOf("quarter" to "Q3", "revenue" to 180000, "profit" to 60000),
    mapOf("quarter" to "Q4", "revenue" to 200000, "profit" to 80000)
)

// Export data to JavaScript
jsExport("salesData", salesData)
jsExport("months", months)
jsExport("productSales", productSales)
jsExport("quarterlyRevenue", quarterlyRevenue)
```

```python
%js

import { salesData, months, productSales, quarterlyRevenue } from '@jupyter';
import * as echarts from 'echarts';

// Create a line chart
const lineChartDom = getCellRoot();
lineChartDom.style = "width: 100%; height: 400px;";

const lineChart = echarts.init(lineChartDom);
const lineOption = {
    title: {
        text: 'Monthly Sales Data',
        left: 'center'
    },
    tooltip: {
        trigger: 'axis'
    },
    xAxis: {
        type: 'category',
        data: months
    },
    yAxis: {
        type: 'value'
    },
    series: [{
        data: salesData,
        type: 'line',
        smooth: true,
        itemStyle: {
            color: '#5470c6'
        }
    }]
};

lineChart.setOption(lineOption);
```

```python
%js

// Create a bar chart
const barChartDom = getCellRoot();
barChartDom.style = "width: 100%; height: 400px;";

const barChart = echarts.init(barChartDom);
const barOption = {
    title: {
        text: 'Quarterly Revenue vs Profit',
        left: 'center'
    },
    tooltip: {
        trigger: 'axis',
        axisPointer: {
            type: 'shadow'
        }
    },
    legend: {
        data: ['Revenue', 'Profit']
    },
    xAxis: {
        type: 'category',
        data: quarterlyRevenue.map(item => item.quarter)
    },
    yAxis: {
        type: 'value'
    },
    series: [
        {
            name: 'Revenue',
            type: 'bar',
            data: quarterlyRevenue.map(item => item.revenue),
            itemStyle: {
                color: '#91cc75'
            }
        },
        {
            name: 'Profit',
            type: 'bar',
            data: quarterlyRevenue.map(item => item.profit),
            itemStyle: {
                color: '#fac858'
            }
        }
    ]
};

barChart.setOption(barOption);
```

```python
%js

// Create a pie chart
const pieChartDom = getCellRoot();
pieChartDom.style = "width: 100%; height: 400px;";

const pieChart = echarts.init(pieChartDom);
const pieData = Object.entries(productSales).map(([name, value]) => ({
    name,
    value
}));

const pieOption = {
    title: {
        text: 'Product Sales Distribution',
        left: 'center'
    },
    tooltip: {
        trigger: 'item',
        formatter: '{a} <br/>{b}: {c} ({d}%)'
    },
    legend: {
        orient: 'vertical',
        left: 'left'
    },
    series: [
        {
            name: 'Sales',
            type: 'pie',
            radius: '50%',
            data: pieData,
            emphasis: {
                itemStyle: {
                    shadowBlur: 10,
                    shadowOffsetX: 0,
                    shadowColor: 'rgba(0, 0, 0, 0.5)'
                }
            }
        }
    ]
};

pieChart.setOption(pieOption);
```
