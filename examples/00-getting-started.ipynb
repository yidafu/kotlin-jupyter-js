# TypeScript Basic Usage Example

This example demonstrates how to use TypeScript magic (`%ts`) in Kotlin Jupyter for type-safe JavaScript development.

## Features
- TypeScript type system
- Import variables from Kotlin
- Type-safe function definitions
- Interface and type definitions


```kotlin
// Prepare data in Kotlin
val userName = "John Doe"
val userAge = 25
val userHobbies = listOf("Programming", "Reading", "Traveling")
val userProfile = mapOf(
    "name" to userName,
    "age" to userAge,
    "hobbies" to userHobbies,
    "isActive" to true
)

// Export variables to JavaScript using jsExport
jsExport("userProfile", userProfile)
jsExport("userHobbies", userHobbies)
```

```kotlin
%ts

// Import data from Kotlin
import { userName, userAge, userProfile, userHobbies } from '@jupyter';

// Define TypeScript interfaces
interface User {
    name: string;
    age: number;
    hobbies: string[];
    isActive: boolean;
}

// Type-safe functions
function greetUser(user: User): string {
    return `Hello, ${user.name}! You are ${user.age} years old.`;
}

function formatHobbies(hobbies: string[]): string {
    return hobbies.join(', ');
}

// Use type assertion to ensure data conforms to interface
const user: User = userProfile as User;

// Call functions and display results
const greeting = greetUser(user);
const hobbiesText = formatHobbies(userHobbies);

// Display results on the page
const root = getCellRoot();
root.innerHTML = `
    <div style="padding: 20px; border: 1px solid #ddd; border-radius: 8px; background: #f9f9f9;">
        <h3>TypeScript Basic Example</h3>
        <p><strong>Greeting:</strong> ${greeting}</p>
        <p><strong>Hobbies:</strong> ${hobbiesText}</p>
        <p><strong>User Status:</strong> ${user.isActive ? 'Active' : 'Inactive'}</p>
        <div style="margin-top: 15px; padding: 10px; background: #e8f4fd; border-radius: 4px;">
            <small>üí° This example demonstrates TypeScript's type safety features and ability to import data from Kotlin</small>
        </div>
    </div>
`;
```
