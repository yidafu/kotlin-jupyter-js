This example demonstrates how to use D3.js library in Kotlin Jupyter to create custom data visualizations.

## Features
- Custom SVG visualizations
- Data binding and manipulation
- Interactive elements
- Scales and axes
- Animations and transitions


```python
// Configure D3.js library
jsConfig {
    d3()
}

// Prepare sample data for visualization
val populationData = listOf(
    mapOf("country" to "China", "population" to 1439323776, "area" to 9596961),
    mapOf("country" to "India", "population" to 1380004385, "area" to 3287263),
    mapOf("country" to "USA", "population" to 331002651, "area" to 9833517),
    mapOf("country" to "Indonesia", "population" to 273523615, "area" to 1904569),
    mapOf("country" to "Pakistan", "population" to 220892340, "area" to 881912),
    mapOf("country" to "Brazil", "population" to 212559417, "area" to 8514877),
    mapOf("country" to "Nigeria", "population" to 206139589, "area" to 923768),
    mapOf("country" to "Bangladesh", "population" to 164689383, "area" to 147570)
)

val temperatureData = listOf(
    mapOf("month" to "Jan", "temp" to 2),
    mapOf("month" to "Feb", "temp" to 4),
    mapOf("month" to "Mar", "temp" to 8),
    mapOf("month" to "Apr", "temp" to 15),
    mapOf("month" to "May", "temp" to 20),
    mapOf("month" to "Jun", "temp" to 25),
    mapOf("month" to "Jul", "temp" to 28),
    mapOf("month" to "Aug", "temp" to 27),
    mapOf("month" to "Sep", "temp" to 22),
    mapOf("month" to "Oct", "temp" to 16),
    mapOf("month" to "Nov", "temp" to 9),
    mapOf("month" to "Dec", "temp" to 4)
)

// Export data to JavaScript
jsExport("populationData", populationData)
jsExport("temperatureData", temperatureData)
```

```python
%js

import { populationData, temperatureData } from '@jupyter';
import * as d3 from 'd3';

// Create a bar chart using D3.js
const container = getCellRoot();
container.innerHTML = '<div id="d3-container"></div>';

const svg = d3.select("#d3-container")
    .append("svg")
    .attr("width", 800)
    .attr("height", 400);

const margin = { top: 20, right: 30, bottom: 40, left: 40 };
const width = 800 - margin.left - margin.right;
const height = 400 - margin.top - margin.bottom;

const g = svg.append("g")
    .attr("transform", `translate(${margin.left},${margin.top})`);

// Create scales
const xScale = d3.scaleBand()
    .domain(populationData.map(d => d.country))
    .range([0, width])
    .padding(0.1);

const yScale = d3.scaleLinear()
    .domain([0, d3.max(populationData, d => d.population)])
    .range([height, 0]);

// Create bars
g.selectAll(".bar")
    .data(populationData)
    .enter().append("rect")
    .attr("class", "bar")
    .attr("x", d => xScale(d.country))
    .attr("width", xScale.bandwidth())
    .attr("y", d => yScale(d.population))
    .attr("height", d => height - yScale(d.population))
    .attr("fill", "#69b3a2");

// Add axes
g.append("g")
    .attr("transform", `translate(0,${height})`)
    .call(d3.axisBottom(xScale))
    .selectAll("text")
    .style("text-anchor", "end")
    .attr("dx", "-.8em")
    .attr("dy", ".15em")
    .attr("transform", "rotate(-45)");

g.append("g")
    .call(d3.axisLeft(yScale));

// Add title
svg.append("text")
    .attr("x", width / 2)
    .attr("y", 20)
    .attr("text-anchor", "middle")
    .style("font-size", "16px")
    .style("font-weight", "bold")
    .text("Population by Country (in millions)");
```

```python
%js

// Create a line chart for temperature data
const lineContainer = getCellRoot();
lineContainer.innerHTML = '<div id="line-chart"></div>';

const lineSvg = d3.select("#line-chart")
    .append("svg")
    .attr("width", 800)
    .attr("height", 400);

const lineG = lineSvg.append("g")
    .attr("transform", `translate(${margin.left},${margin.top})`);

// Create line scales
const xLineScale = d3.scaleBand()
    .domain(temperatureData.map(d => d.month))
    .range([0, width])
    .padding(0.1);

const yLineScale = d3.scaleLinear()
    .domain(d3.extent(temperatureData, d => d.temp))
    .range([height, 0]);

// Create line generator
const line = d3.line()
    .x(d => xLineScale(d.month) + xLineScale.bandwidth() / 2)
    .y(d => yLineScale(d.temp))
    .curve(d3.curveMonotoneX);

// Add the line
lineG.append("path")
    .datum(temperatureData)
    .attr("fill", "none")
    .attr("stroke", "#e74c3c")
    .attr("stroke-width", 3)
    .attr("d", line);

// Add circles for data points
lineG.selectAll(".dot")
    .data(temperatureData)
    .enter().append("circle")
    .attr("class", "dot")
    .attr("cx", d => xLineScale(d.month) + xLineScale.bandwidth() / 2)
    .attr("cy", d => yLineScale(d.temp))
    .attr("r", 5)
    .attr("fill", "#e74c3c");

// Add axes
lineG.append("g")
    .attr("transform", `translate(0,${height})`)
    .call(d3.axisBottom(xLineScale));

lineG.append("g")
    .call(d3.axisLeft(yLineScale));

// Add title
lineSvg.append("text")
    .attr("x", width / 2)
    .attr("y", 20)
    .attr("text-anchor", "middle")
    .style("font-size", "16px")
    .style("font-weight", "bold")
    .text("Monthly Temperature Trends");
```
