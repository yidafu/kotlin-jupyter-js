## Step 1: Generate Structured Data in Kotlin

Create data classes and collections representing real-world entities.


````kotlin
// Define data classes
// Note: @Serializable annotation requires %use serialization to be executed first (see cell above)
// This annotation is required for jsExport() to serialize data classes to JSON
// Without this annotation, the data cannot be exported to JavaScript environment
@Serializable
data class Employee(
    val id: Int,
    val name: String,
    val department: String,
    val salary: Double,
    val joinDate: String
)

// Generate employee data
val employees = listOf(
    Employee(1, "Alice Johnson", "Engineering", 95000.0, "2020-01-15"),
    Employee(2, "Bob Smith", "Marketing", 75000.0, "2019-06-20"),
    Employee(3, "Carol White", "Sales", 70000.0, "2021-03-10"),
    Employee(4, "David Brown", "Engineering", 105000.0, "2018-11-05"),
    Employee(5, "Eva Davis", "HR", 65000.0, "2022-01-08"),
    Employee(6, "Frank Miller", "Engineering", 98000.0, "2020-09-12")
)

// âš ï¸ IMPORTANT: Custom classes with @Serializable annotation MUST use jsExport() to be accessible in JavaScript
// Default import (import { variable } from '@jupyter') does NOT support custom classes
// Only basic types (String, Number, Boolean, Map, List, etc.) can be imported directly
// For @Serializable classes, you MUST explicitly export them using jsExport():
jsExport("employees", employees)

println("âœ… Structured data created successfully")
println("Created: ${employees.size} employees")
````

## Step 3: Render Employee Table

Create a styled HTML table to display employee information using Tailwind CSS.


````kotlin
%js

import { employees } from '@jupyter';

const container = getContainer();

// Calculate statistics
const totalEmployees = employees.length;
const avgSalary = employees.reduce((sum, emp) => sum + emp.salary, 0) / totalEmployees;
const maxSalary = Math.max(...employees.map(emp => emp.salary));
const departments = [...new Set(employees.map(emp => emp.department))];

container.innerHTML = `
    <div class="p-5 font-sans">
        <h2 class="text-slate-700 border-b-2 border-blue-500 pb-2.5 mb-4 text-2xl font-semibold">
            ðŸ‘¥ Employee Directory
        </h2>
        
        <div class="bg-gray-100 p-4 rounded-lg my-4">
            <h3 class="text-slate-600 mt-0 mb-2 text-lg font-semibold">Statistics</h3>
            <p class="mb-1"><strong>Total Employees:</strong> ${totalEmployees}</p>
            <p class="mb-1"><strong>Average Salary:</strong> $${avgSalary.toLocaleString('en-US', {minimumFractionDigits: 2, maximumFractionDigits: 2})}</p>
            <p class="mb-1"><strong>Highest Salary:</strong> $${maxSalary.toLocaleString('en-US', {minimumFractionDigits: 2, maximumFractionDigits: 2})}</p>
            <p class="mb-0"><strong>Departments:</strong> ${departments.join(', ')}</p>
        </div>
        
        <div class="overflow-x-auto mt-5 shadow-md rounded-lg">
            <table class="w-full border-collapse">
                <thead>
                    <tr class="bg-gradient-to-r from-indigo-500 to-purple-600 text-white">
                        <th class="p-3 text-left border border-gray-300">ID</th>
                        <th class="p-3 text-left border border-gray-300">Name</th>
                        <th class="p-3 text-left border border-gray-300">Department</th>
                        <th class="p-3 text-right border border-gray-300">Salary</th>
                        <th class="p-3 text-left border border-gray-300">Join Date</th>
                    </tr>
                </thead>
                <tbody>
                    ${employees.map((emp, index) => `
                        <tr class="${index % 2 === 0 ? 'bg-white' : 'bg-gray-50'}">
                            <td class="p-2.5 border border-gray-300">${emp.id}</td>
                            <td class="p-2.5 border border-gray-300 font-medium">${emp.name}</td>
                            <td class="p-2.5 border border-gray-300">
                                <span class="bg-blue-500 text-white px-2 py-1 rounded text-xs">
                                    ${emp.department}
                                </span>
                            </td>
                            <td class="p-2.5 border border-gray-300 text-right text-green-600 font-semibold">
                                $${emp.salary.toLocaleString('en-US', {minimumFractionDigits: 2, maximumFractionDigits: 2})}
                            </td>
                            <td class="p-2.5 border border-gray-300">${emp.joinDate}</td>
                        </tr>
                    `).join('')}
                </tbody>
            </table>
        </div>
    </div>
`;

console.log('Employee table rendered:', employees);
````
