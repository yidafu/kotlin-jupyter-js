# Vis Network

```kotlin
val nodes = arrayOf(
    mapOf("id" to 1, "label" to "node 1"),
    mapOf("id" to 2, "label" to "node 2"),
    mapOf("id" to 3, "label" to "node 3"),
    mapOf("id" to 4, "label" to "node 4"),
    mapOf("id" to 5, "label" to "node 5"),
)
val edges = arrayOf(
    mapOf("from" to 1, "to" to 3),
    mapOf("from" to 1, "to" to 2),
    mapOf("from" to 2, "to" to 4),
    mapOf("from" to 2, "to" to 5),
    mapOf("from" to 3, "to" to 3),
)
```

# Vis Graph3d

Offical Graph3d demo. <https://visjs.github.io/vis-graph3d/examples/graph3d/01_basics.html>

```kotlin

fun custom(x: Double, y: Double): Double {
  return Math.sin(x / 50) * Math.cos(y / 50) * 50 + 50;
}
  
val dataList = mutableListOf<Map<String, Double>>()

val steps = 50; // number of datapoints will be steps*steps
val axisMax = 314;
val axisStep = axisMax / steps;
var x = 0.0;
var counter: Double = 0.0;
while (x < axisMax) {
    var y = 0.0;
    while (y < axisMax) {
        val v = custom(x, y);
        counter += 1.0;
        dataList.add(mapOf( "id" to counter,  "x" to x, "y" to y, "z" to v, "style" to v ))
        y += axisStep;
    }
    x += axisStep;
}
```

```kotlin
%js
import { dataList } from '@jupyter';

import { Graph3d, DataSet } from "https://esm.sh/vis-graph3d@6.0.6/standalone";


var data = new DataSet(dataList)


var options = {
    width: "600px",
    height: "600px",
    style: "surface",
    showPerspective: true,
    showGrid: true,
    showShadow: false,
    keepAspectRatio: true,
    verticalRatio: 0.5,
};

// Instantiate our graph object.
var container = getCellRoot('100%', '400px');
var graph = new Graph3d(container, data, options);
```
